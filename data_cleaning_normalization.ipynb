```python
policy=pd.read_csv("policies.csv")
policy.reset_index(drop=True)
policy.head()
#policy.set_index("date",inplace=True)
#policy.index=pd.to_datetime(policy.index)
policy.drop(policy.columns[0], axis=1, inplace=True)
policy.columns
```

```python
c=[8,10,12,14,16,18,20,22,23,25,26,28,29,30,32,34]
for i in c:
    print(policy.columns[i-1])
```

## For each policy, divide it by the maximum possible and make it a percentage

```python
for i in c:
    my_max = policy[policy.columns[i-1]].loc[policy[policy.columns[i-1]].idxmax()]
    policy[policy.columns[i-1]]=policy[policy.columns[i-1]] / my_max *100
    policy[policy.columns[i-1]]=policy[policy.columns[i-1]].apply(lambda x: x if np.isnan(x) else int(x))
```
